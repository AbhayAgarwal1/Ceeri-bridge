## Year 1 (Data Extraction)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import csv
import numpy as np
import csv
from scipy.fftpack import fft
x = []
y = []
a = []
b = []
c = []
d = []
f = open('firsty.txt')
csv_f = csv.reader(f)
i=0;
for row in csv_f:
  x.append((row[0]))
  y.append((row[1]))
  a.append((row[2]))
  b.append((row[3]))
  c.append((row[4]))
  d.append((row[5]))
  i = i+1
  
  

from sklearn.preprocessing import scale

ys = scale(y)
ab = scale(a)
bs = scale(b)
cs = scale(c)
ds = scale(d)


#FFT
#number of sample points
N = 41701
#frequency of signal 
T = 1//200
#create x-axis for time length of signal
x = np.linspace(0, N*T, N)
#create array that corresponds to values in signal
#y = df
#perform FFT on signal

'''plt.plot(x,y, label='Loaded from file!')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()

plt.plot(x,ys, label='Loaded from file!')

plt.plot(x,ab, label='Loaded from file!')
plt.plot(x,bs, label='Loaded from file!')
plt.plot(x,cs, label='Loaded from file!')


plt.title('Data from the CSV File')
plt.xticks([])
plt.yticks([])
plt.xlabel('time')
plt.ylabel('acc (m/s2)')

plt.show()
'''
```

# pillar 1 (fft)

```python
yf = fft(ys)
#create new x-axis: frequency from signal
xf = np.linspace(0.0, 200/2, N//2)
#plot results
plt.plot(xf, yf[0:N//2], label = 'signal')
plt.grid()
plt.xlabel('Frequency (days)')
plt.ylabel(r'Spectral Amplitude')
plt.legend(loc=1)
plt.savefig('rsam_2016_2017_snippet_fft_firstresult.jpg')
plt.show()
```
